Nhiệm vụ 5 – Pivot Table + Stack/Unstack

```python
#5.1 Pivot: Thời gian xử lý trung bình
  #đọc dữ liệu
import pandas as pd

df = pd.read_csv("../task04_merg e/full_feedback_merged.csv")

#theo department × category
pivot_time = pd.pivot_table(
    df,
    index='department',
    columns='category',
    values='resolution_hours',
    aggfunc='mean'
)

pivot_time
```

```python
# 5.2 Tạo pivot số lượng phản ánh theo area × category. 
pivot_count = pd.pivot_table(
    df,
    index='area',
    columns='category',
    values='feedback_id',
    aggfunc='count',
    fill_value=0
)

pivot_count
```

```python
# 5.3 Thực hành MultiIndex – stack / unstack
#Stack (columns → index)
stacked_time = pivot_time.stack()
stacked_time.head()
```

```python
#Unstack (index → columns)
unstacked_time = stacked_time.unstack()
unstacked_time.head()
```

```python
#5.4 Tìm department × category có thời gian xử lý cao nhất
max_time = stacked_time.idxmax()
max_value = stacked_time.max()

max_time, max_value
```
