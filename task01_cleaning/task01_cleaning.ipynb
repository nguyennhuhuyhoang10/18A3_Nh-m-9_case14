Nhiệm vụ 1 – Đọc & làm sạch dữ liệu

```python
#1.1 Làm sạch citizen_id, full_name và area trong citizen_info.csv. 
import pandas as pd
import numpy as np
# Đọc dữ liệu
citizen_df = pd.read_csv("../data_raw/citizen_info.csv")
# Chuẩn hoá citizen_id: xoá khoảng trắng, chuyển về string
citizen_df['citizen_id'] = citizen_df['citizen_id'].astype(str).str.strip()
# Làm sạch full_name: xoá khoảng trắng dư, viết hoa chữ cái đầu
citizen_df['full_name'] = (
    citizen_df['full_name']
    .astype(str)
    .str.strip()
    .str.title()
)

# Làm sạch area: xoá khoảng trắng, chuẩn hoá chữ thường
citizen_df['area'] = (
    citizen_df['area']
    .astype(str)
    .str.strip()
    .str.lower()
)

```

```python
#Chuẩn hóa category, channel, department và submit_time trong feedback_log.csv. 
feedback_df = pd.read_csv("../data_raw/feedback_log.csv")
# Chuẩn hoá text
for col in ['category', 'channel', 'department']:
    feedback_df[col] = (
        feedback_df[col]
        .astype(str)
        .str.strip()
        .str.lower()
    )

# Chuẩn hoá submit_time về datetime
feedback_df['submit_time'] = pd.to_datetime(
    feedback_df['submit_time'],
    errors='coerce'
)
feedback_df.to_csv("../data_clean/feedback_log_clean.csv", index=False)
```

```python

#1.4Chuẩn hóa resolved_time, status và resolution_hours trong resolution_time.csv. 

resolution_df = pd.read_csv("../data_raw/resolution_time.csv")
# status
resolution_df['status'] = (
    resolution_df['status']
    .astype(str)
    .str.strip()
    .str.lower()
)

# resolved_time → datetime
resolution_df['resolved_time'] = pd.to_datetime(
    resolution_df['resolved_time'],
    errors='coerce'
)
# resolution_hours → numeric
resolution_df['resolution_hours'] = pd.to_numeric(
    resolution_df['resolution_hours'],
    errors='coerce'
)

#1.5.Xử lý resolution_hours bị sai kiểu (text, khoảng trắng, None) và các giá trị bất thường (giá trị âm)

# Loại bỏ giá trị âm
resolution_df.loc[
    resolution_df['resolution_hours'] < 0,
    'resolution_hours'
] = np.nan
resolution_df.to_csv("../data_clean/resolution_time_clean.csv", index=False)
```

```python
# Lưu file sạch
citizen_df.to_csv("../data_clean/citizen_info_clean.csv", index=False)
feedback_df = pd.read_csv("../data_raw/feedback_log.csv")
```
